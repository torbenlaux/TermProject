# **Benefits of Active Modes of Transportation**
    
</div>

Active transportation has numerous benefits beyond its positive environmental impact. For instance, walking and cycling can improve physical health, reduce traffic congestion, and promote social interactions among residents. Furthermore, cities with robust active transportation infrastructure tend to be more livable, attractive, and vibrant, which can enhance economic growth and social well-being.

Overall, studying the active transportation initiatives and outcomes of Montreal, Toronto, and Vancouver can provide valuable lessons for other cities seeking to become more sustainable and livable.

Here are some benefits of active transportation:

<b> Health Benefits: </b>

* Active transportation helps individuals meet their daily physical activity requirements, reducing the risk of chronic diseases such as heart disease, obesity, and diabetes. According to the World Health Organization, insufficient physical activity is the fourth leading risk factor for global mortality (World Health Organization (WHO), 2010).
 
* Active transportation, such as cycling and walking, can improve community and individual well-being. A study showed that cyclists reported their commute as the most pleasant activity of the day more often than drivers. Spending time in automobile traffic and congestion can cause stress and reduce work productivity. Noise pollution from cars is associated with mental health issues like depression (EcoPlan International, 2011, p. 17).
 
<b> Environmental Benefits: </b>

* Active transportation reduces greenhouse gas emissions and air pollution, contributing to a healthier environment. According to a report by the Intergovernmental Panel on Climate Change, the transportation sector is responsible for approximately 23% of global energy-related greenhouse gas emissions (Sims R, at all. Transport. In: Climate Change 2014: Mitigation of Climate Change. Contribution of Working Group III to the Fifth Assessment Report of the Intergovern- mental Panel on Climate Change).

<b> Economic Benefits: </b>

* Active transportation can save individuals money on transportation costs and reduce the demand for car-centric infrastructure, freeing up resources for other public services(Transport Canada, 2016).

<b> Social Benefits: </b>

* Car-dependent communities can increase social and physical distances between neighbors, workplaces, and amenities, reducing opportunities for social contact and weakening neighborliness and trust. A study conducted in San Francisco showed that heavy traffic has a negative impact on the stress and social behavior of a neighborhood. Residents on streets with heavy traffic were more likely to withdraw from the street, whereas those on streets with light traffic had more friends and acquaintances. Designing communities for active transportation can foster community interaction and encourage a greater sense of community (EcoPlan International, 2011, p. 18).

### <b> Clean up Data 

```python
# Read the CSV file
cma_data = pd.read_csv('cma_data.csv', encoding='iso-8859-1')

#Rename columns for better comprehension
cma_data.rename(columns={"GEO UID":"GEO ID","ProvCode": "Province Code","ProvNom":"Province Name","CDcode":"City Code","CDnom":"City","DAnom":"DDANOM","Total":"Total","Car":"Motorized","Public":"Public", "Walk": "Walk", "Bike":"Bike","Other":"Delete"}, inplace = True)

#Delete all unwanted columns
cma_data.drop('Delete', inplace = True, axis=1)
cma_data.drop('DDANOM', inplace = True, axis=1)
cma_data.drop('Province Name', inplace = True, axis=1)
cma_data.drop('City Code', inplace = True, axis=1)


#Add Walk and Bike columns together, to create the "Active Transport" category

cma_data['Active'] = cma_data['Walk'] + cma_data['Bike']


#Delete Walk and Bike, since Active will be conserved to represent the two
cma_data.drop('Walk', inplace = True, axis=1)
cma_data.drop('Bike', inplace = True, axis=1)


cma_data
```

# Part 2 - Getting The Data Ready

* Add a new column called City Name : Greater Montreal Area, Greater Toronto Area, andGreater Vancouver Area


* Sum of the rows for each city


#### <b> Separate the dataframe per CMA </b>

```python
grandMtl = cma_data[cma_data['Province Code'] == 24]
grandTor = cma_data[cma_data['Province Code'] == 35]
grandVan = cma_data[cma_data['Province Code'] == 59]

#Now that we know that Province Code 24 represents Quebec and, in this case, all of the Greater Montreal Area- 
# we will delete the City column and do a sum of all the rows. This will show all values for GMA

grandMtl.to_csv('montreal_data.csv', index=False)
grandTor.to_csv('toronto_data.csv', index=False)
grandVan.to_csv('vancouver_data.csv', index=False)


# Re-enter the CSV files 
grandMtl = pd.read_csv('montreal_data.csv')
grandTor = pd.read_csv('toronto_data.csv')
grandVan = pd.read_csv('vancouver_data.csv')


grandMtl.drop('City', inplace = True, axis=1)
grandTor.drop('City', inplace = True, axis=1)
grandVan.drop('City', inplace = True, axis=1)

#Remove NAs and nulls

grandMtl.dropna(inplace=True)
grandTor.dropna(inplace=True)
grandVan.dropna(inplace=True)
```

#### <b> Make a sum of all rows per Province - so there is only one row </b>

```python
#MONTREAL
sumMtl = grandMtl.sum(axis=0)
sumMtl = sumMtl.to_frame().T

#TORONTO
sumTor = grandTor.sum(axis=0)
sumTor = sumTor.to_frame().T

#VANCOUVER
sumVan = grandVan.sum(axis=0)
sumVan = sumVan.to_frame().T

#Remove GEO ID, Province ID and add column with just CMA names
sumMtl['City'] = 'Montreal'
sumTor['City'] = 'Toronto'
sumVan['City'] = 'Vancouver'


sumMtl.drop('GEO ID', inplace = True, axis=1)
sumMtl.drop('Province Code', inplace = True, axis=1)
sumTor.drop('GEO ID', inplace = True, axis=1)
sumTor.drop('Province Code', inplace = True, axis=1)
sumVan.drop('GEO ID', inplace = True, axis=1)
sumVan.drop('Province Code', inplace = True, axis=1)

#Test to see if it worked
# sumVan
# sumMtl
sumTor
```

## <b> Lets define a function that takes the city name as an argument and returns a bar plot of the trip data for that city.
### This way it will be easy to plot each CMA later. </b>

```python
def plot_city_trips(city, color):
    # read the CSV file
    df = pd.read_csv('all_cities.csv')

    # select only the specified city to display
    df = df[df['City'] == city]

    # group by city and sum the values
    df = df.groupby('City').sum()

    # get the maximum values for each category
    max_values = df[['Motorized', 'Public', 'Active', 'Total']].max()

    # create a horizontal bar plot with the specified color
    max_values.plot(kind='barh', color=color)

    # add labels and title
    plt.xlabel('Number of Trips')
    plt.ylabel('Category')
    plt.title('{} Trips by Category'.format(city))

    # show the plot
    plt.show()
    
plot_city_trips('Montreal', 'lightgreen')
plot_city_trips('Vancouver', 'pink')
plot_city_trips('Toronto', 'lightblue')
```

## Now lets make a fuction that will plot them all at once

```python
def plot_city_trips(city, ax):
    # read the CSV file
    df = pd.read_csv('all_cities.csv')

    # select only the specified city to display
    df = df[df['City'] == city]

    # group by city and sum the values
    df = df.groupby('City').sum()

    # get the maximum values for each category
    max_values = df[['Motorized', 'Public', 'Active', 'Total']].max()

    # create a horizontal bar plot with different colors
    colors = ['lightblue', 'lightgreen', 'pink', 'yellow']
    max_values.plot(kind='barh', color=colors, ax=ax)

    # add labels and title
    ax.set_xlabel('Number of Trips')
    ax.set_ylabel('Category')
    ax.set_title('{} Trips by Category'.format(city))

# create a figure and three subplots
fig, axs = plt.subplots(3, 1, figsize=(6, 12))

# plot the three cities on their respective subplots
plot_city_trips('Montreal', axs[0])
plot_city_trips('Vancouver', axs[1])
plot_city_trips('Toronto', axs[2])

# adjust the layout and padding
plt.tight_layout(pad=3.0)

# show the plot
plt.show()
```

## Lets make a new fuction to better understand the data. We will do this so all data is out of 100%.

```python
def plot_city_trips(cities):
    # read the CSV file
    df = pd.read_csv('all_cities.csv')

    # select only the specified cities to display
    df = df[df['City'].isin(cities)]

    # group by city and sum the values
    df = df.groupby('City').sum()

    # calculate the percentages for each category
    df_percent = df.apply(lambda x: x/x.sum()*100, axis=1)

    # create a horizontal bar plot with different colors for each city
    ax = df_percent[['Motorized', 'Public', 'Active', 'Total']].plot(kind='barh', stacked=True, figsize=(20, 8), 
                                                                     color=['lightblue', 'pink', 'lightgreen', 'yellow'])
    for i in ax.containers:
        ax.bar_label(i, label_type='center', fontsize=10, padding=3)
        
    ax.set_xlabel('Percentage of Trips')
    ax.set_ylabel('Canadian CMA')
    ax.set_title('Trips by Category at 100% for {}'.format(', '.join(cities)))

    # save the plot as a PNG image with a unique name based on the list of cities
    filename = '_'.join(cities) + '_trips_by_category.png'
    plt.savefig(filename, dpi=300)

    # show the plot
    plt.show()

plot_city_trips(['Montreal', 'Vancouver', 'Toronto'])
```
